# Radioactivity

- Theory of Radioativity 
    - Radioactivity is the process where unstable atomic nuclei release energy and particles, such as alpha, beta, and gamma radiation, to become more stable.
    - This decay is a spontaneous and random event, but its rate is constant for a given isotope, a property known as half-life.
    - The emission of radiation can be a natural process or can occur in human-made materials, and it has numerous applications, from medicine to smoke detectors. 

- What is radioactivity?
    - Unstable nuclei: Atoms with an unstable nucleus have an excess of internal energy, often due to an imbalance of neutrons and protons.

    - Spontaneous decay: To become stable, these unstable atoms undergo a process called radioactive decay, emitting energy in the form of particles or electromagnetic waves.

    - Change of element: This decay process converts the original atom into a different isotope or even a different chemical element. 



- Formula: 

A= - {dN}//{dt}

A	=	total activity

N	=	number of particles

t	=	time

- (A) Calculate the decay constant for cobalt-60 (T1/2 = 5.26 years) in units of month−1.
- (B) What will be the activity of a 5,000-Ci 60Co source after 4 years?

```python
# Program: Calculate Decay Constant and Activity for Cobalt-60 using NumPy and Pandas

import numpy as np
import pandas as pd

##(A) Calculate the decay constant for cobalt-60 (T1/2 = 5.26 years) in units of month−1.
# Given data
T_half_years = 5.26     # half-life in years
T_half_months = T_half_years * 12      # convert half-life to months

# Calculate decay constant (λ) in month⁻¹
lamda = np.log(2) / T_half_months


#Result_A = {"Half-Life (years)": [T_half_years],
   # "Half-Life (months)": [round(T_half_months, 2)],
    #"Decay Constant λ (month⁻¹)": [round(lamda, 5)]}


Result_A = {"Half-Life (years)": [T_half_years],
    "Half-Life (months)": [(T_half_months)],
    "Decay Constant λ (month⁻¹)": [(lamda)]}

# Display results Sectoion A 
print("\n--- Radioactivity Calculation for Cobalt-60 Section (A)---")
print(Result_A)


# (B) What will be the activity of a 5,000-Ci 60Co source after 4 years?

A0 = 5000               # initial activity in curies (Ci)
t_years = 4              # time elapsed in years

# Convert units

t_months = t_years * 12                # convert time to months

# Calculate decay constant (λ) in month⁻¹
lamda = np.log(2) / T_half_months

# Calculate remaining activity using exponential decay formula
A = A0 * np.exp(-lamda * t_months)




# Create a DataFrame to display results neatly
Result_B = {
    "Initial Activity A₀ (Ci)": [A0],
    "Time Elapsed (months)": [t_months],
    "Remaining Activity A (Ci)": [round(A, 3)]}


# Display results
print("\n--- Radioactivity Calculation for Cobalt-60 for Section (B)---")
print(Result_B)





```
