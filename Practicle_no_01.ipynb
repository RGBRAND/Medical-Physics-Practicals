# Write a Python Program using operators to perform oprations on variables 

- Theory 

- - Variables are essentially containers that store different types of data, such as numbers (integers, floats), text (strings), or more complex structures like lists or dictionaries.

 - Python supports various built-in data types for variables, including the fundamental types mentioned:
     - Integers (int): These represent whole numbers, positive or negative, without any decimal component.
     - Floats (float): These represent numbers with a decimal point, allowing for fractional values.
     - Strings (str): These represent sequences of characters, used for text. Strings are enclosed in single or double quotes.
     - Booleans (bool): These represent logical values, either True or False. They are often used for conditional logic and comparisons.



- Arithmetic Operators in Python:
- - Python provides a comprehensive set of operators for performing mathematical calculations on numerical variables:

     - Addition (+): Adds two operands.
     - Subtraction (-): Subtracts the second operand from the first.
     - Multiplication (*): Multiplies two operands.
     - Division (/): Divides the first operand by the second, always returning a float.
     - Floor Division (//): Divides and rounds down to the nearest whole number (integer). 
     - Exponentiation (`: `):** Raises the first operand to the power of the second. 


```python
# Define arithmetic variables
num1 = 25
num2 = 7

# Perform arithmetic operations
sum_result = num1 + num2
difference_result = num1 - num2
product_result = num1 * num2
division_result = num1 / num2
floor_division_result = num1 // num2
power_result = num1 ** 2
Squar_root = num1**(0.5)

# Print the results
print(f"Sum: {sum_result}")
print(f"Difference: {difference_result}")
print(f"Product: {product_result}")
print(f"Division: {division_result}")
print(f"Floor Division: {floor_division_result}")
print(f"Power: {power_result}")
print(f'Square root: {Squar_root}')
```

Description:

  - In Python, every value or variable has a data type that defines the kind of data it holds (e.g., integer, string, float, list, etc.).
To find the data type of any variable, we use the type() function.


- The type() function tells what kind of data a variable holds.

  - Python automatically assigns the data type based on the value stored.

- Common data types include:

    - int â†’ Integer numbers

    - float â†’ Decimal numbers

    - str â†’ Strings (text)

    - bool â†’ True/False values

    - list, tuple, dict, set â†’ Collection types



```python
# Program to find out data types of different variables

a = 10             # Integer
b = 3.14           # Float
c = "Hello"        # String
d = True           # Boolean
e = [1, 2, 3]      # List
f = (4, 5, 6)      # Tuple
g = {"name": "Sujay", "age": 25}   # Dictionary

print("Data type of a:", type(a))
print("Data type of b:", type(b))
print("Data type of c:", type(c))
print("Data type of d:", type(d))
print("Data type of e:", type(e))
print("Data type of f:", type(f))
print("Data type of g:", type(g))
```
